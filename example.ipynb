```python
from qiskit import *    
```

```python
qr = QuantumRegister(2) ## Build a quantum registry for 2 qbits
```

```python
cr = ClassicalRegister(2) ## Builds a quantum registry for 2 qbits
```

```python
circuit = QuantumCircuit(qr, cr)
```

```python
circuit.draw(output="mpl")
```

```python
circuit.h(qr[0])
```

```python
circuit.draw(output="mpl")
```

```python
circuit.cx(qr[0], qr[1]) ## This is going to enttagle the qbits together
```

```python
circuit.draw(output="mpl")
```

```python
circuit.measure(qr,cr)
```

```python
circuit.draw(output="mpl")
```
